```python
import geemap
import ee
geemap.ee_initialize()
```

**Question 1**

```python
Map=geemap.Map(center=[40,-100], zoom=5)
dataset = ee.FeatureCollection('USGS/WBD/2017/HUC04')
style = {'color': '0000ffff', 'width': 2, 'lineType': 'solid', 'fillColor': 'FF000000'}
Map.addLayer(dataset.style(**style), {}, 'US Water Areas')
Map.add_text('Water Counties by Javier Martinez', fontsize=20, position= 'bottomright')
Map
```

**Question 3**

```python
Map=geemap.Map()
dataset = ee.FeatureCollection('USGS/WBD/2017/HUC04')
area= dataset.filter(ee.Filter.eq('name', 'Wabash'))
dem= ee.Image("USGS/3DEP/10m")
elevation=dem.select('elevation')
vis= {
    'min': 0.0,
    'max': 950,
    'palette': 'terrain'
}
clip = dem.clipToCollection(area)
style = {'color': '0000ffff', 'width': 3, 'lineType': 'solid', 'fillColor': 'FF000000'}
Map.addLayer(area.style(**style), {}, 'Wabash County')
Map.centerObject(area, zoom=7)
Map.addLayer(clip,vis, 'elevation')
Map.add_colorbar(vis, label='Elevation (m)')
Map.add_text('Wabash DEM by Javier Martinez', fontsize=20, position= 'bottomright')
Map
```

**Question 4**

```python
Map=geemap.Map(center=[40, -100], zoom=5)
states = ee.FeatureCollection("TIGER/2018/States").filter(ee.Filter.eq('NAME', 'Utah'))
collection01 = (
    ee.ImageCollection('USGS/NLCD_RELEASES/2019_REL/NLCD').filter(ee.Filter.eq('system:index', '2001')).select('landcover')
    .filterBounds(states)
    .first()
)
img01=collection01.clipToCollection(states)
collection19 = (
    ee.ImageCollection('USGS/NLCD_RELEASES/2019_REL/NLCD').filter(ee.Filter.eq('system:index', '2019')).select('landcover')
    .filterBounds(states)
    .first()
)
img19=collection19.clipToCollection(states)
left_layer = geemap.ee_tile_layer(img01, {}, 'NLCD 2001')
right_layer = geemap.ee_tile_layer(img19, {}, 'NLCD 2019')
Map.split_map(left_layer, right_layer, add_close_button=True)
Map.center_object(states, zoom=7)
Map.add_legend(title='NLCD Land Cover', builtin_legend='NLCD')
Map.add_text('Utah split map by Javier Martinez', fontsize=20, position= 'bottomright')
Map
```
