```python
#Para hacer operaciones a una lista que siempre son las mismas (multiplicar, sumar...) 
#utilizar la función .map(arg....)
def computeSquares(number):
    return ee.Number(number).pow(2).subtract(1)

#diseñamos la función que queramos y después aplicamos .map()
squares = myList.map(computeSquares)
squares
```
